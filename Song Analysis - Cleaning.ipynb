```python
import re
import string

def clean_text_r1(text):
    text = text.lower()
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
    text = re.sub('\w*\d\w*', '', text)
    return text

round1 = lambda x: clean_text_r1(x)
```

```python
dylan_df['lyrics'] = dylan_df['lyrics'].apply(round1)
```

```python
dylan_df.head()
```

```python
def clean_text_r2(text):
    text = re.sub('\n',' ', text)
    text = re.sub('\r',' ', text)
    return text

round2 = lambda x : clean_text_r2(x)
```

```python
dylan_df['lyrics'] = dylan_df['lyrics'].apply(round2)
```

```python
def stem_text(text):
    processed_tokens = []
    tokens = nltk.word_tokenize(text)
    for t in tokens:
        t = ps.stem(t)
        processed_tokens.append(t)
    return " ".join(processed_tokens)

round3 = lambda x: stem_text(x)
```
